# Simon's Nilearn Extras
Showing the basic functions of simons_nilearn_extras. 
As said in the readme, these functions use the code from nilearn very heavily, so definitely give them all the credit!

```python
# Load packages:
import matplotlib.pyplot as plt
import simons_nilearn_extras as ne
from nilearn import plotting, datasets

%matplotlib inline
```

```python
# Use MNI template here:
mni = datasets.MNI152_FILE_PATH
```

```python
# Define a couple of seed regions:
seeds = [[19, 10, 54],  [-12, 21, 1]]
# Create masks for these regions:
voi_masks = ne.create_spherical_vois(mni, seeds, radius=8)
```

```python
# Get the labels accordingly:
labels = ne.coordinate_label(seeds)
```

```python
# Plot regions
fig, axes = plt.subplots(2,1, figsize=(20,10))

for voi, label, ax in zip(voi_masks, labels, axes.ravel()):
    plotting.plot_roi(voi, title=label, display_mode='z', cut_coords=1, axes=ax, figure=fig)
```
